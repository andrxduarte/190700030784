```python
#Definição que exclui os indicadores dos dias da coluna e cria uma nova tabela com a somas de sales por month
def month_years_sales(data, time=['monthly','years']):
    data = data.copy()
    if time == 'monthly':
        data.date = data.date.apply(lambda x: str(x)[:-3])
    else:
        data.date = data.date.apply(lambda x: str(x)[:4])
        
    data = data.groupby('date')['sales'].sum().reset_index()
    data.date = pd.to_datetime(data.date)
    
    return data
```

```python
df_m = month_years_sales(df_s, 'monthly')
df_m.to_csv('./monthly_data.csv')
```

Como verificar a estacionariedade de uma série temporal?

Diz-se que um TS é estacionário se suas propriedades estatísticas , como média e variância, permanecem constantes ao longo do tempo . Mas por que é importante? A maioria dos modelos TS trabalha com a suposição de que o TS é estacionário. Intuitivamente, podemos perceber que se um TS tem um determinado comportamento ao longo do tempo, há uma probabilidade muito alta de que ele siga o mesmo no futuro. Além disso, as teorias relacionadas às séries estacionárias são mais maduras e fáceis de implementar em comparação com as séries não estacionárias.

A estacionariedade é definida usando um critério muito estrito. No entanto, para fins práticos, podemos assumir que a série é estacionária se ela tiver propriedades estatísticas constantes ao longo do tempo, ou seja, Os seguintes:

* média constante
* variância constante
* uma autocovariância que não depende do tempo

>Existem muitos outros modelos para prever séries temporais, como modelos de média móvel ponderada ou modelos de média móvel integrada autorregressiva (ARIMA). Alguns deles exigem que você primeiro remova a tendência e a sazonalidade. Por exemplo, se você está estudando o número de usuários ativos em seu site e está crescendo 10% a cada mês, você deve remover essa tendência da série temporal. Depois que o modelo é treinado e começa a fazer previsões, você precisa adicionar a tendência de volta para obter as previsões finais. Da mesma forma, se você estiver tentando prever a quantidade de protetor solar vendida todo mês, provavelmente observará uma forte sazonalidade: como vende bem todo verão, um padrão semelhante se repetirá todos os anos. Você teria que remover essa sazonalidade da série temporal, por exemplo, calculando a diferença entre o valor em cada intervalo de tempo e o valor um ano antes (essa técnica é chamada de diferenciação). Novamente, depois que o modelo é treinado e faz previsões, você precisa adicionar o padrão sazonal de volta para obter as previsões finais.

Visualizando a plotagem dos gráficos observamos que eles não são estácionários.

Para torná-lo estacionário, precisa calcular a diferença entre as vendas de cada mês e adicioná-la nos dados como uma nova coluna.

```python
def sales_time(data):
    """Time interval of dataset:"""

    data.date = pd.to_datetime(data.date)
    n_of_days = data.date.max() - data.date.min()
    n_of_years = int(n_of_days.days / 365)
    
    print(f'Dias: {n_of_days.days}\nAnos: {n_of_years}\nMeses: {12 * n_of_years}')
```

# Determinando Séries Temporais Estacionárias

O princípio subjacente é modelar ou estimar a tendência e a sazonalidade da série e removê-los da série para obter uma série estacionária. Então técnicas de previsão estatística podem ser implementadas nesta série. A etapa final seria converter os valores previstos na escala original, aplicando de volta as restrições de tendência e sazonalidade.

Esta função de ajuda salvará o erro médio quadrático (RMSE) e o erro médio absoluto (MAE) das nossas previsões para comparar o desempenho dos nossos modelos.

![](https://4.bp.blogspot.com/-wG7IbjTfE6k/XGUvqm7TCVI/AAAAAAAAAZU/vpH1kuKTIooKTcVlnm1EVRCXLVZM9cPNgCLcBGAs/s1600/formula-MAE-MSE-RMSE-RSquared.JPG)

```python
def time_plot(data, x_col, y_col, title):
    fig, ax = plt.subplots(figsize = (15,8))
    sns.lineplot(x_col, y_col, data = data, ax = ax, color = 'darkblue', label='Total de Vendas')
    
    s_mean = data.groupby(data.date.dt.year)[y_col].mean().reset_index()
    s_mean.date = pd.to_datetime(s_mean.date, format='%Y')
    sns.lineplot((s_mean.date + datetime.timedelta(6*365/12)), y_col, data=s_mean, ax=ax, color='red', label='Média de Vendas')   
    
    ax.set(xlabel = 'Anos',
           ylabel = 'Vendas',
           title = title)
```
